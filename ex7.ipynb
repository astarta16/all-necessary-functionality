## შექმენი შემდეგი სტრუქტურები oop პრინციპების გამოყენებით

თითოეული uppercase კლასს ჰქონდეს შესაბამისი lowercase ფუნქცია, რომელსაც ექნება return "კლასის სახელი"
```python
class T:
    def t(self):
        return "class T"
```

````python
#__start__
class A:
  def a(self):
    return "class A"
class B(A):
  def b(self):
    return "class B"
class D:
  def d(self):
    return "class D"
class C(A,D):
  def c(self):
    return "class C"
class E(C,D):
  def e(self):
    return "class E"

#__end__
````
