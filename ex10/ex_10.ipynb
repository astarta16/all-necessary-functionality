შეფასების კრიტერიუმები

* ამოცანის სწორად ამოხსნა 1 ქულა

* მეთოდების, ოპერატორების სწორად გამოყენება (კოდის წერა) 1 ქულა

* ხარვეზების დაჭერა და გამოსწორება (debuging) 2 ქულა

* დამატებითი რესურსების გამოყენება (stackoverflow, google search) 1 ქულა


```python
import json
file = open("base.txt")
result = {}
for i in file:
# delete unnecessary things
    i = i\
            .replace(" ", "_")\
            .replace(","," ")\
            .replace("#"," ")\
            .replace('"', "")\
            .replace("\n", "")   
    x = i.find(" ")
    result[i[0:x]] = i[x:].split()
#  delete line with no team
del result["Team"]
# convert str to int
for i,k in result.items():
    lst = []
    for a in k:
        try:
            lst.append(int(a))
        except:
            if "." in a:
                a = float(a)
            elif a == "True":
                a = 1
            elif a == "False":
                a = 0
            lst.append(a)
    result[i] = lst
best_team = {}
worst_team = {}
# find best and worst teams
for k, v in result.items():
    b = round(v[1]/v[0]*100,2) 
    best_team[k] = b
    if v[2] > v[1] and v[2] > v[3]:
        worst_team[k] = v
# sort 
best_team = dict(sorted(best_team.items(),key=lambda item: item[1])[-3:])
worst_team = dict(sorted(worst_team.items(),key=lambda item: item[1]))
result = {}
# save in result
result["best_team"] = best_team
result["worst_team"] = list(worst_team.keys())
# create file
with open("result.json", "w") as write_file:
    json.dump(result, write_file, indent=1)
```
