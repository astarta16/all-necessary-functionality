<h2>ამოცანა 1 (გენერატორები)</h2>

* დააგენერირეთ n დან m მდე რიცხვების კვადრატები
* გენერატორით ამოიღეთ თითოეულიდან მეოთხე ხარისხის ფესვი  
* დაამრგვალეთ თითოეული ელემენტი მეასედებამდე
* დადაიყვანეთ სტრინგში შედეგი  და გამოყავით space ებით

genik(50,60) -> '2.66 2.67 2.69 2.7 2.71 2.72 2.74 2.75 2.76 2.77'

genik(5,6) -> '1.5'

genik(10,61) -> '1.78 1.82 1.86 1.9 1.93 1.97 2.0 2.03 2.06 2.09 2.11 2.14 2.17 2.19 2.21 2.24 2.26 2.28 2.3 2.32 2.34 2.36 2.38 2.4 2.41 2.43 2.45 2.47 2.48 2.5 2.51 2.53 2.55 2.56 2.58 2.59 2.6 2.62 2.63 2.65 2.66 2.67 2.69 2.7 2.71 2.72 2.74 2.75 2.76 2.77 2.78'



````python
#__start_
def genik(n, m):
  x = [i for i in range (n, m)]
  y = [str(round(i**0.25, 2)) for i in x]
  return " ".join(y)


#__end_
````

<h2>ამოცანა 2 (დეკორატორები)</h2>

წინა ამოცანაში დაწერილი ფუნქცია დაადეკორირეთ მოცემული კლასის მეშვეობით
* დეკორირების პრინციპი: 
    ზედა ზღვარი გაზარდეთ 10 ით

```python
class decorator:
    def __init__(self,a,b):
        self.a = a
        self.b = b
        
    def sum_nums(self):
        return self.a + self.b

```



````python
#__start_
class decorator:
  def __init__(self, *function):
    self.function = function[0]

  def __call__(self, a, b):
    print(self.function(a, b+10))

  def sum_nums(self, a, b):
    return a + b

@decorator
def genik(n, m):
  x = [i for i in range (n, m)]
  y = [str(round(i**0.25, 2)) for i in x]
  return " ".join(y)



#__end_
````

<h2>ამოცანა 3 (ფუნქცია დეკორატორები)</h2>

წინა ამოცანაში დაწერილი დეკორატორი გადააკეთეთ ფუნქცია დეკორატორად

* დეკორირების პრინციპი: 
    ზედა და ქვედა ზღვარი შეატრიალეთ
    მიღებული ქვედა ზღვარი გაყავით 2 ზე


````python
#__start_
def decorator(function):
  def wrapper(a, b):
    print(function(int(b/2), a))
  return wrapper 

@decorator
def genik(n, m):
  x = [i for i in range (n, m)]
  y = [str(round(i**0.25, 2)) for i in x]
  return " ".join(y)
genik(50,60)

#__end_
````
